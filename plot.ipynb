# ERCA - Group 8

This notebook present the surface temperature evolution since 1980, compared to the average temperature over 20 years (1958-1977).

Data are taken from Goddard Institute for Space Studies (GISS)

Source: https://data.giss.nasa.gov/cgi-bin/gistemp/stdata_show_v4.cgi?id=FRM00007591&ds=14&dt=1

For Embrum Station

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# load data
df = pd.read_csv("data.txt", sep='\\s+', skiprows=2)
# replace 999.9 --> nan
arr = np.where(np.isclose(df.values, 999.90), np.nan, df.values)
df = pd.DataFrame(arr, index=df.index, columns=df.columns)
```

```python
# compute average annual temperature between 1958 and 1977
Tmean=np.average(df.metANN[:20])
```

## Plot annual trend 

```python
plt.plot(df.YEAR[20:],df.metANN[20:]-Tmean)
plt.xlabel('Year')
plt.ylabel('Temperature (Celsius)')
plt.title('Surface temperature evolution at Embrun')
```
